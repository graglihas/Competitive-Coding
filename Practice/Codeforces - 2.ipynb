## Tram (116A)

```
capacity=0
ppl=0
for i in range(int(input())):
  a,b = map(int,input().split(' '))
  ppl = ppl-a+b
  capacity = max(capacity,ppl)
print(capacity)
```

## Head or Tails (242A)

```
x,y,a,b = map(int,input().split(' '))
count=0;flag=1
comb=[]
if a<b:
  if x>=b:
    a=b
  else:
    print(0)
    flag=0
if flag is not 0:
  for i in range(a,x+1):
    for j in range(b,min(i+1,y+1)):
      if i is j:
        continue
      else:
        comb.append([i,j])
        count+=1
  print(count)
  if count>0:
    for i in comb:
      print(*i)
```

## Marks (152A)

```
n,m = map(int,input().split(' '))
marks=[[0]*n for _ in range(m)]
for i in range(n):
  s=input()
  for j in range(m):
    marks[m-j-1][i]=int(s[m-j-1])    
best=[0]*n
for i in range(m):
  ele = max(marks[i])
  for j in range(n):
    if marks[i][j] is ele:
      best[j]=1
print(sum(best))
```
