# FCTRL2

```
# cook your dish here
factorial=[]
def fact(num):
    if len(factorial)>=num:
        return factorial[num-1]
    else:
        if num ==1:
            f=1
        else:
            f = num*fact(num-1)
        factorial.append(f)
        return f
        
for i in range(int(input())):
    number = int(input())
    print(fact(number))
```

```
# cook your dish here
for i in range(int(input())):
    l,r = map(int,input().split(' '))
    pretty =0
    while l%10!=0:
        if l%10 == 2 or l%10 ==3 or l%10 ==9:
            pretty = pretty +1
        l=l+1
    while r%10!=0:
        if r%10 == 2 or r%10 ==3 or r%10 ==9:
            pretty = pretty +1
        r=r-1
    pretty = pretty + (int((r-l)/10)*3)
    print(pretty)
```
