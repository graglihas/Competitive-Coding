## Xenia and Ringroad (339B)

```
n,m =map(int,input().split(' '))
a = list(map(int,input().split(' ')))
time=0
current=1
for i in range(m):
  if a[i]==current:
    continue
  if a[i]>current:
    time=time+(a[i]-current)
  if a[i]<current:
    time=time+n-current+1+(a[i]-1)
  current=a[i]
print(time)
```

## Lucky Division (122A)

```
def lucky_num(a):
  for i in range(len(a)):
    if a[i] =='4' or a[i]=='7':
      continue
    else:
      return 0
  return 1


num=int(input())
if lucky_num(str(num))==1:
  print('YES')
else:
  flag=0
  for i in range(1,num):
    j=str(i)
    if num%i is 0 and lucky_num(j) is 1:
      flag=1
      print('YES')
      break
  if flag==0:
    print('NO')
```

## Twins(160A)

```
n = int(input())
a=list(map(int,input().split(' ')))
a.sort()
total =sum(a)

coins=0;amt=0
for i in range(n-1,-1,-1):
  amt+=a[i]

  coins+=1
  if amt>(total/2):
    break
print(coins)
```

## Chatroom (58A)

```
s=input()
i=0
j=len(s)
hello=[0]*5
while i<j and s[i]!='h':
  i+=1
if i<j and s[i]=='h':
  hello[0]=1
i+=1
while i<j and s[i]!='e':
  i+=1
if i<j and s[i]=='e':
  hello[1]=1
i+=1
while i<j and s[i]!='l':
  i+=1
if i<j and s[i]=='l':
  hello[2]=1
i+=1
while i<j and s[i]!='l':
  i+=1
if i<j and s[i]=='l':
  hello[3]=1
i+=1
while i<j and s[i]!='o':
  i+=1
if i<j and s[i]=='o':
  hello[4]=1
if sum(hello)==5:
  print('YES')
else:
  print('NO')
```

## K-string (219A)

```
k=int(input())
s=input()
c=[0]*26
for i in range(len(s)):
  index = ord(s[i])-97
  c[index]+=1
flag=1
ch=''

for i in range(26):
  if c[i]==0:
    continue
  if c[i]%k==0:
    ch=ch+(chr(i+97)*int(c[i]/k))
  else:
    flag=0
    break
if flag==1:
  print(ch*k)
else:
  print(-1)
```

## Dragons (230A)

```
s,n = map(int,input().split(' '))
d=[]
for i in range(n):
  x,y = map(int,input().split(' '))
  d.append([x,y])
d=sorted(d,key=lambda x: x[0])
flag=1
for i in range(n):
  if d[i][0]<s:
    s=s+d[i][1]
  else:
    flag=0
    break
if flag==0:
  print('NO')
else:
  print('YES')
```

## Puzzles(337A)

```
n,m = map(int,input().split(' '))
f=list(map(int,input().split(' ')))
f.sort()
r=float('inf')
for i in range(m-n+1):
  r=min(r,(f[i+n-1]-f[i]))
print(r)
```
