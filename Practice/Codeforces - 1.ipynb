# Dividing Orange (244A)

```
n,k =map(int,input().split(' '))
a = list(map(int,input().split(' ')))
s = [[0]*n for _ in range(k)]
#print(s)
for j in range(k):
  s[j][0] = a[j]
  #print(s)
nk=1
for i in range(k):
  for j in range(1,n):
    while nk in a:
      nk = nk +1
    s[i][j]=nk
    nk=nk+1
for i in range(k):
  for j in range(n):
    print(s[i][j],end=' ')
```

# Lunch Rush (276A)

```
n,k=map(int,input().split(' '))
fun = float('-inf')
for i in range(n):
  f,t=map(int,input().split(' '))
  if t<=k:
    fun = max(f,fun)
  else:
    fun = max((f-t+k),fun)
print(int(fun))
```

# I_love_%username% (155A)

```
n = int(input())
score = list(map(int,input().split(' ')))
min = score[0];max = score[0]
amaz = 0
for i in range(1,n):
  if score[i]>max:
    max = score[i]
    amaz +=1
  if score[i]<min:
    min = score[i]
    amaz +=1
print(amaz)
```

# cakeminator (330A)

```
a,b = map(int,input().split(' '))
cake=[]
for i in range(a):
  cake.append(input())
eat =0
for i in range(a):
  if 'S' not in cake[i] and '.' in cake[i]:
    eat += cake[i].count('.')
    cake[i]=cake[i].replace('.','x')
#print(cake)
for i in range(b):
  if 'S'not in [row[i] for row in cake] and '.' in [row[i] for row in cake]:
    eat += [row[i] for row in cake].count('.')
print(eat)
```

# insomnia cure (148A)

```
k=int(input())
l=int(input())
m=int(input())
n=int(input())
d=int(input())
dragon = ['.']*d
l = set([k,l,m,n,k*l,k*m,k*n,l*m,l*n,m*n,k*l*m,k*l*n,k*m*n,l*m*n,k*l*m*n])
for i in l:
  for j in range(i-1,d,i):
    dragon[j]='*'
print(dragon.count('*'))
```

# vanya and card (401A)

```
n,x=map(int,input().split(' '))
card = list(map(int,input().split(' ')))
want = sum(card)
if want ==0:
  print(0)
else:
  if abs(want)<=x:
    print(1)
  else:
    count=0
    while abs(want)>x:
      want =abs(want)-x
      count+=1
    print(count+1)
```
