# MEET

```
# cook your dish here
def to_24(time_12):
    if time_12[5]=='A':
        time_12 = int(time_12[:5])
    else:
        time_12 = int(time_12[:5])+1200
    if (time_12 >= 1200 and time_12 <=1259) or (time_12 >= 2400 and time_12 <=2459) :
        time_12 = time_12 - 1200
    return time_12
        
for i in range(int(input())):
    meet = to_24(input().replace(':',''))
    attend = ''
    for j in range(int(input())):
        time = input().replace(':','')
        start = to_24(time[:7])
        end = to_24(time[8:])
        if meet >= start and meet <=end:
            attend =attend +'1'
        else:
            attend = attend +'0'
    print(attend)
```

# TEAMNAME

```
# cook your dish here
for i in range(int(input())):
    n = int(input())
    s = input().split(' ')
    good_name = 0
    for j in range(n):
        for k in range(j+1,n,1):
            a = s[j][0]+s[k][1:]
            b = s[k][0]+s[j][1:]
            if a not in s and b not in s:
                good_name = good_name+2
    print(good_name)
```
